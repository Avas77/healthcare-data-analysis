```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import title
```

```python
patient_csv = pd.read_csv("dataset/patients_data_clean.csv")
patient_csv
```

```python
type(patient_csv["Visit_Date"][0])
```

```python
patient_csv["Visit_Date"] = pd.to_datetime(patient_csv["Visit_Date"])
```

```python
type(patient_csv["Visit_Date"][0])
```

```python
patient_csv.set_index("Visit_Date", inplace=True)
patient_csv
```

```python
patient_csv_resampled = patient_csv.resample("Y").count()
```

```python
patient_csv_resampled
```

```python
plt.figure(figsize=(12, 5))
plt.plot(patient_csv_resampled.index, patient_csv_resampled["Patient_ID"])
plt.title("Trend Analysis of Patients by Year")
plt.xlabel("Date")
plt.ylabel("Number of Patients")
plt.grid(True)
plt.show()
```

```python
patient_csv_resampled_month = patient_csv.resample("M").count()
```

```python
patient_csv_resampled_month
```

```python
plt.figure(figsize=(12, 5))
plt.plot(patient_csv_resampled_month.index, patient_csv_resampled_month["Patient_ID"])
plt.title("Trend Analysis of Patients by Month")
plt.xlabel("Date")
plt.ylabel("Number of Patients")
plt.grid(True)
plt.show()
```

```python
monthly_ticks = pd.date_range(start=patient_csv_resampled_month.index.min(), end=patient_csv_resampled_month.index.max(), freq="M")
monthly_labels = [date.strftime('%b%Y') for date in monthly_ticks]
plt.figure(figsize=(12, 5))
plt.plot(patient_csv_resampled_month.index, patient_csv_resampled_month["Patient_ID"])
plt.title("Trend Analysis of Patients by Month")
plt.xticks(monthly_ticks, monthly_labels, rotation=90)
plt.xlabel("Date")
plt.ylabel("Number of Patients")
plt.grid(True)
plt.show()
```

```python
target_month = 2

df_target_month = patient_csv_resampled_month[patient_csv_resampled_month.index.month == target_month]
plt.figure(figsize=(12, 5))
plt.plot(df_target_month.index, df_target_month["Postcode"])
plt.title("Trend Analysis of Patients by Month of February")
plt.xlabel("Date")
plt.ylabel("Number of Visits")
plt.grid(True)
plt.show()
```
