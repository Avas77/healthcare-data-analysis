```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
patient_df = pd.read_csv('dataset\patients_data_clean.csv')
patient_df.head()
```

```python
numeric_df = patient_df.drop(["Visit_Date", "Patient_ID", "Gender", "Diagnosis"], axis=1)
```

```python
plt.figure(figsize=(12,5))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
